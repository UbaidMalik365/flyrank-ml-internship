## 4. The unit of analysis, as a real dataframe

*Load your lane's slice and show it: one row = one what?*


The unit of analysis is **one web page**. Each row in the dataset represents a single web page and includes observable features such as content type, impressions, click-through rate (CTR), average search position, and trend direction. For this project, the target label **is_declining_label** is derived from the `trend_direction` column, where pages with a trend of **down** are labeled as declining. This page-level information will be used to rank pages that should be prioritized for content refresh.

```python
import pandas as pd

df = pd.read_csv("data/raw/content_refresh_anonymized.csv")

# Target label
df["is_declining_label"] = (
    df["trend_direction"].str.lower() == "down"
).astype(int)

print("Dataset Shape:", df.shape)

print("\nUnit of analysis: One row represents one web page.\n")

display(df[[
    "content_id",
    "content_type",
    "impressions_90d",
    "ctr",
    "avg_position",
    "trend_direction",
    "is_declining_label"
]].head(10))
```

## 5. Why ML beats a fixed rule here

*What makes the pattern too messy for an if-statement?*


A fixed rule uses only a small number of conditions, such as page age or impressions.

Machine learning can evaluate many observable signals together, including impressions, click-through rate, average position, content age, and freshness. This allows the model to recognize more complex patterns and produce a better-ranked list of pages for review.

The model supports decision-making, while the final refresh decision remains with the content team.

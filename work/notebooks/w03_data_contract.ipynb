## Fields: Feature / Label / Context / Excluded

### Features
- impressions_90d
- ctr
- avg_position
- content_age_days
- days_since_last_update

These are observable signals that are available before making a content refresh decision.

### Label
- is_declining_label

This label is created from the `trend_direction` column. Pages with a trend of **down** are labeled as declining.

### Context
- content_type
- trend_direction

These fields provide additional information about each page but are not the primary features used for prediction.

### Excluded
- content_id

This field is only an identifier and does not help the model make predictions.

```python
import pandas as pd

df = pd.read_csv("data/raw/content_refresh_anonymized.csv")

# Create label
df["is_declining_label"] = (
    df["trend_direction"].str.lower() == "down"
).astype(int)

print("Features:")
print([
    "impressions_90d",
    "ctr",
    "avg_position",
    "content_age_days",
    "days_since_last_update"
])

print("\nLabel:")
print("is_declining_label")

print("\nContext:")
print(["content_type", "trend_direction"])

print("\nExcluded:")
print(["content_id"])
```

## Data Limits

This dataset contains observable search and content signals only. It does not prove why a page declined or guarantee that refreshing a page will improve performance.

The analysis is observational and supports decision-making, not causal conclusions. Other external factors, such as search engine updates, competition, or seasonality, are not fully captured in the dataset.

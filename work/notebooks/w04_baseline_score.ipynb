## 2. Build the ranked queue (writes the CSV)

*Code the score, rank everything, write work/outputs/baseline_action_score.csv.*

## Build the Ranked Queue

I created a simple baseline scoring rule that prioritizes pages with high visibility and stale content. Pages with more impressions and at least 180 days since the last update receive higher scores.

The output includes:
- A baseline score
- A reason code
- An action label

The ranked queue is written to `work/outputs/baseline_action_score.csv`.

```python
import pandas as pd
import os

# Load dataset
df = pd.read_csv("data/raw/content_refresh_anonymized.csv")

# Baseline score
df["baseline_score"] = (
    df["impressions_90d"] *
    (df["days_since_last_update"] >= 180).astype(int)
)

# Reason code
df["reason_code"] = "STALE_VISIBLE_PAGE"

# Action label
df["action"] = "Refresh Review"

# Sort highest score first
queue = df.sort_values("baseline_score", ascending=False)

# Create output folder
os.makedirs("work/outputs", exist_ok=True)

# Save CSV
queue.to_csv(
    "work/outputs/baseline_action_score.csv",
    index=False
)

print("CSV written successfully!")
print(queue[[
    "content_id",
    "baseline_score",
    "reason_code",
    "action"
]].head(10))
```

## 3. Top-20 review

*For each of the top 20: action, reason code, confidence note, and what would make it wrong.*

## Top-20 Review

The table below shows the top 20 pages selected by the baseline rule. These pages are recommended for a content refresh because they have high visibility and have not been updated recently.

The confidence is medium because the recommendation is based on observable signals only. A recommendation could be wrong if a page is seasonal, intentionally unchanged, or already scheduled for an update.

```python
import pandas as pd

# Load dataset
df = pd.read_csv("data/raw/content_refresh_anonymized.csv")

# Baseline score
df["baseline_score"] = (
    df["impressions_90d"] *
    (df["days_since_last_update"] >= 180).astype(int)
)

# Reason code
df["reason_code"] = "STALE_VISIBLE_PAGE"

# Action
df["action"] = "Refresh Review"

# Confidence
df["confidence"] = "Medium"

# What could make it wrong?
df["what_would_make_it_wrong"] = (
    "Seasonal traffic or page already planned for update."
)

# Top 20
top20 = df.sort_values(
    "baseline_score",
    ascending=False
).head(20)

display(top20[[
    "content_id",
    "baseline_score",
    "action",
    "reason_code",
    "confidence",
    "what_would_make_it_wrong"
]])
```

## 4. Weak picks + leakage check

*Which picks look wrong and why? Confirm no product flags or future windows leaked in.*

## Weak Picks + Leakage Check

Some pages selected by the baseline rule may not actually need a content refresh. For example, a page may have high impressions but be seasonal, already updated, or intentionally unchanged.

I confirmed that no label-derived features or future information were used in the baseline score. The rule only uses observable signals that are available before making the decision:
- impressions_90d
- days_since_last_update

Therefore, the baseline does not suffer from data leakage.

```python
import pandas as pd

# Load dataset
df = pd.read_csv("data/raw/content_refresh_anonymized.csv")

# Create baseline score
df["baseline_score"] = (
    df["impressions_90d"] *
    (df["days_since_last_update"] >= 180).astype(int)
)

# Show some low-scoring pages
weak_picks = df.sort_values("baseline_score").head(10)

print("Lowest scoring pages:")
display(
    weak_picks[
        [
            "content_id",
            "impressions_90d",
            "days_since_last_update",
            "baseline_score"
        ]
    ]
)

print("\nLeakage Check:")
print("No future-window columns or label-derived features were used.")
```
